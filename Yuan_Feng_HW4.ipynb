```python
import pandas as pd
import numpy as np
%load_ext sql
%sql sqlite://
```

```python
song = pd.read_csv("spotify_songs.csv")
```

```python
song=song.drop(['Unnamed: 0'], axis=1)
song
```

```python
track_id_album_id = song[["track_id","track_album_id"]]
track_id_playlist_id = song[["track_id","playlist_id"]]
playlist_id_album_id = song[["playlist_id","track_album_id"]]
```

```python
album_id_album_name = song[["track_album_id","track_album_name"]]
album_id_album_date = song[["track_album_id","track_album_release_date"]]
```

```python
playlist_id_name= song[["playlist_id","playlist_name"]]
playlist_id_genre= song[["playlist_id","playlist_genre"]]
playlist_id_subgenre= song[["playlist_id","playlist_subgenre"]]
```

```python
track_id_track_name = song[["track_id","track_name"]]
track_id_track_artist = song[["track_id","track_artist"]]

track_id_more_details = song[["track_id","track_popularity","danceability","energy","key","loudness","mode","speechiness","acousticness","instrumentalness","liveness","valence","tempo","duration_ms"]]
```

```python
%%sql

DROP TABLE IF EXISTS track_id_album_id;
DROP TABLE IF EXISTS track_id_playlist_id;
DROP TABLE IF EXISTS album_id_album_name;
DROP TABLE IF EXISTS album_id_album_date;
DROP TABLE IF EXISTS playlist_id_name;
DROP TABLE IF EXISTS playlist_id_genre;
DROP TABLE IF EXISTS playlist_id_subgenre;
DROP TABLE IF EXISTS track_id_track_name;
DROP TABLE IF EXISTS track_id_track_artist;
DROP TABLE IF EXISTS track_id_more_details;
DROP TABLE IF EXISTS playlist_id_album_id;
```

```python
%sql PERSIST track_id_album_id;
%sql PERSIST track_id_playlist_id;
%sql PERSIST album_id_album_name;
%sql PERSIST album_id_album_date;
%sql PERSIST playlist_id_name;
%sql PERSIST playlist_id_genre;
%sql PERSIST playlist_id_subgenre;
%sql PERSIST track_id_track_name;
%sql PERSIST track_id_track_artist;
%sql PERSIST track_id_more_details;
%sql PERSIST playlist_id_album_id;
```

```python
%%sql 
create table temp as
select distinct(p.playlist_id)
from track_id_playlist_id as p 
left join track_id_more_details as t
on t.track_id = p.track_id
where t.instrumentalness > 0.5;
```

```python
%%sql  

select distinct(p.playlist_name)
from temp
left join playlist_id_name as p 
on p.playlist_id = temp.playlist_id
```
